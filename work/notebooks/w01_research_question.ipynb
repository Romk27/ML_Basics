## 4. Careful words: what I can and can't claim

This project aims to provide decision support by identifying pages that may be at higher risk of future decline based on historical content and performance signals. The results can show patterns and improve prioritization for review, but they cannot prove that any feature causes a page to decline. Likewise, the model does not predict Google's ranking algorithm or guarantee that refreshing a page will improve its performance. Its purpose is to estimate future risk using historical observations.
